```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=50, scale=10, size=500)

# 使用 seaborn 绘制直方图
sns.histplot(data, bins=30, kde=True, color="blue")

# 显示图表
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram with Seaborn")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据（正态分布）
data = np.random.normal(loc=50, scale=10, size=100)

# 使用 seaborn 绘制箱线图
sns.boxplot(data=data, color="lightblue")

# 设置标题和标签
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.title("Boxplot with Seaborn")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据（正态分布）
data = np.random.normal(loc=50, scale=10, size=100)

# 绘制小提琴图
sns.violinplot(data=data, color="lightblue", inner="quartile")

# 设置标题和标签
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.title("Violin Plot with Seaborn")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
np.random.seed(42)
x = np.random.rand(100) * 10   # x 轴数据（0~10的随机数）
y = 2 * x + np.random.normal(scale=2, size=100)  # y = 2x + 噪声

# 绘制散点图
sns.scatterplot(x=x, y=y, color="blue", alpha=0.7)

# 设置标题和标签
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot with Seaborn")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
np.random.seed(42)
x = np.random.rand(100) * 10  # x 轴数据（0~10 的随机数）
y = 2 * x + np.random.normal(scale=2, size=100)  # y = 2x + 噪声

# 绘制散点图 + 线性回归直线
sns.regplot(x=x, y=y, ci=95, line_kws={"color": "red"})

# 设置标题和标签
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression with Seaborn")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 生成一个随机数据集
np.random.seed(42)
data = np.random.rand(10, 10)  # 10x10 矩阵数据
df = pd.DataFrame(data, columns=[f"Var{i+1}" for i in range(10)])

# 计算相关性矩阵
corr_matrix = df.corr()

# 绘制热力图
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# 设置标题
plt.title("Heatmap of Correlation Matrix")

# 显示图表
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 生成时间序列数据
np.random.seed(42)
dates = pd.date_range(start="2024-01-01", periods=100, freq="D")  # 100 天
values = np.cumsum(np.random.randn(100))  # 生成随机波动的累加数据

# 创建 DataFrame
df = pd.DataFrame({"Date": dates, "Value": values})

# 绘制时间序列图
plt.figure(figsize=(10, 5))
sns.lineplot(x="Date", y="Value", data=df, marker="o", color="b", label="Time Series")

# 设置标题和格式
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Plot")
plt.xticks(rotation=45)  # 旋转 x 轴标签
plt.legend()
plt.grid()

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 创建分类数据
np.random.seed(42)
categories = ["A", "B", "C", "D", "E"]
values = np.random.randint(10, 100, size=5)  # 生成 10~100 之间的随机整数

# 创建 DataFrame
df = pd.DataFrame({"Category": categories, "Value": values})

# 绘制柱状图
plt.figure(figsize=(8, 5))
sns.barplot(x="Category", y="Value", data=df, palette="viridis")

# 添加标题和标签
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Category Bar Plot")

# 显示数值
for index, value in enumerate(values):
    plt.text(index, value + 2, str(value), ha="center", fontsize=12)

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 生成随机数据集
np.random.seed(42)
df = pd.DataFrame({
    "Feature1": np.random.rand(50) * 10,
    "Feature2": np.random.rand(50) * 20,
    "Feature3": np.random.rand(50) * 30,
    "Feature4": np.random.rand(50) * 40,
    "Category": np.random.choice(["A", "B", "C"], size=50)  # 分类变量
})

# 生成多变量关系矩阵（散点图矩阵）
sns.pairplot(df, hue="Category", palette="coolwarm", diag_kind="kde")

# 显示图表
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 创建示例数据集
np.random.seed(42)
df = pd.DataFrame({
    "Feature1": np.random.rand(100) * 10,
    "Feature2": np.random.rand(100) * 20,
    "Category": np.random.choice(["A", "B", "C"], size=100)  # 分类变量
})

# 创建 FacetGrid
g = sns.FacetGrid(df, col="Category", col_wrap=3, height=4)

# 绘制每个子图的散点图
g.map(sns.scatterplot, "Feature1", "Feature2")

# 设置标题
g.set_axis_labels("Feature1", "Feature2")
g.set_titles("Category: {col_name}")

# 显示图表
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.layouts import row
import numpy as np

# 准备数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建第一个图表
p1 = figure(title="正弦函数")
p1.line(x, y1, color="navy")

# 创建第二个图表
p2 = figure(title="余弦函数")
p2.line(x, y2, color="firebrick")

# 并排显示
show(row(p1, p2))
```

```python
from bokeh.plotting import figure, show

# 数据
fruits = ['苹果', '香蕉', '梨', '橙子']
counts = [5, 3, 4, 2]

# 创建图表
p = figure(x_range=fruits, title="水果数量", toolbar_location=None)

# 添加柱状图
p.vbar(x=fruits, top=counts, width=0.9)

# 设置y轴从0开始
p.y_range.start = 0

# 显示图表
show(p)
```
