```python
import matplotlib.pyplot as plt

# 生成斐波那契数列的函数
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence

# 设置斐波那契数列的长度
n = 20

# 获取斐波那契数列
fib_sequence = fibonacci(n)

# 绘制斐波那契数列
plt.plot(range(n), fib_sequence, marker='o', linestyle='-', color='b')
plt.title('Fibonacci Sequence')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Set mean and covariance matrix
mean = [0, 0]  # Mean
cov = [[1, 0.5], [0.5, 1]]  # Covariance matrix

# Create a grid
x, y = np.mgrid[-3:3:0.01, -3:3:0.01]
pos = np.dstack((x, y))  # Convert the grid into coordinate pairs

# Compute the multivariate normal distribution's probability density function
rv = multivariate_normal(mean, cov)
z = rv.pdf(pos)

# Visualization
plt.figure(figsize=(10, 8))
plt.contourf(x, y, z, levels=50, cmap='viridis')  # Plot contour plot with high quality
plt.colorbar(label='Probability Density')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Gaussian Distribution')
plt.show()
```

```python
# 定义分数区间和对应的评级
grade_dict = {
    (90, 100): "A",
    (80, 89): "B",
    (70, 79): "C",
    (60, 69): "D",
    (0, 59): "F"
}

def get_grade(score):
    for range_tuple, grade in grade_dict.items():
        if range_tuple[0] <=    score <= range_tuple[1]:
            return grade

# 测试函数
score = 90
print(f"Score: {score}, Grade: {get_grade(score)}")
```
