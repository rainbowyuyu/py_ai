```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数 f(x) 和它的导数
def f(x):
    return x**2

def df(x):
    return 2*x

# 创建x值的范围
x = np.linspace(-5, 5, 400)
y = f(x)
dy = df(x)

# 绘制原始函数和导数
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制原始函数 f(x) = x^2
ax.plot(x, y, label=r'$f(x) = x^2$', color='blue')

# 绘制第一次求导 f'(x) = 2x
ax.plot(x, dy, label=r"$f'(x) = 2x$", color='red', linestyle='--')

# 逐步绘制二次导数 f''(x) = 2
d2y = np.full_like(x, 2)  # 二阶导数是常数 2
ax.plot(x, d2y, label=r"$f''(x) = 2$", color='green', linestyle=':')

# 设置图例
ax.legend(loc='upper left')

# 添加标题和标签
ax.set_title("Function and its Derivatives")
ax.set_xlabel("x")
ax.set_ylabel("y")

# 显示图形
plt.grid(True)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_function_and_partial_derivatives_3d(func_expr, x_range=(-10, 10), y_range=(-10, 10), num_points=100):
    x, y = sp.symbols('x y')
    f = func_expr  # 输入的函数表达式
    fx = sp.diff(f, x)  # 对x求偏导数
    fy = sp.diff(f, y)  # 对y求偏导数

    # 转换为数值函数
    f_np = sp.lambdify((x, y), f, 'numpy')
    fx_np = sp.lambdify((x, y), fx, 'numpy')
    fy_np = sp.lambdify((x, y), fy, 'numpy')

    # 定义x和y的取值范围
    x_vals = np.linspace(x_range[0], x_range[1], num_points)
    y_vals = np.linspace(y_range[0], y_range[1], num_points)
    X, Y = np.meshgrid(x_vals, y_vals)

    # 计算函数值及偏导数值
    Z = f_np(X, Y)
    Z_fx = fx_np(X, Y)
    Z_fy = fy_np(X, Y)

    # 绘制3D图像
    fig = plt.figure(figsize=(18, 6))

    # 绘制原函数的三维图像
    ax1 = fig.add_subplot(1, 3, 1, projection='3d')
    ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    ax1.set_title(f'f(x, y) = {sp.pretty(f)}')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('z')

    # 绘制∂f/∂x的三维图像
    ax2 = fig.add_subplot(1, 3, 2, projection='3d')
    ax2.plot_surface(X, Y, Z_fx, cmap='plasma', edgecolor='none')
    ax2.set_title(f"∂f/∂x = {sp.pretty(fx)}")
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_zlabel('z')

    # 绘制∂f/∂y的三维图像
    ax3 = fig.add_subplot(1, 3, 3, projection='3d')
    ax3.plot_surface(X, Y, Z_fy, cmap='cividis', edgecolor='none')
    ax3.set_title(f"∂f/∂y = {sp.pretty(fy)}")
    ax3.set_xlabel('x')
    ax3.set_ylabel('y')
    ax3.set_zlabel('z')

    plt.tight_layout()
    plt.show()

# 示例：绘制函数 f(x, y) = x**2 + y**2 及其偏导数
x, y = sp.symbols('x y')
plot_function_and_partial_derivatives_3d(x**2 + y**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数 f(x, y) 和它的偏导数
def f(x, y):
    return x**2 + y**2

def df_dx(x, y):
    return 2*x

def df_dy(x, y):
    return 2*y

# 创建网格
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
Z_dx = df_dx(X, Y)
Z_dy = df_dy(X, Y)

# 绘制函数 f(x, y) 的等高线图
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# 绘制 f(x, y) 的等高线
cp1 = ax[0].contour(X, Y, Z, 20, cmap='viridis')
ax[0].set_title(r'$f(x, y) = x^2 + y^2$')
fig.colorbar(cp1, ax=ax[0])

# 绘制对 x 的偏导数 df/dx 的等高线图
cp2 = ax[1].contour(X, Y, Z_dx, 20, cmap='viridis')
ax[1].set_title(r'$\frac{\partial f}{\partial x} = 2x$')
fig.colorbar(cp2, ax=ax[1])

# 绘制对 y 的偏导数 df/dy 的等高线图
cp3 = ax[2].contour(X, Y, Z_dy, 20, cmap='viridis')
ax[2].set_title(r'$\frac{\partial f}{\partial y} = 2y$')
fig.colorbar(cp3, ax=ax[2])

# 设置标签和标题
for axes in ax:
    axes.set_xlabel('x')
    axes.set_ylabel('y')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, cos, series

# 定义符号变量
x, y = symbols('x y')

# 定义二元函数 f(x, y)
f = sin(x) * cos(y)

# 计算泰勒级数展开（以 (0, 0) 为中心，展开到 4 阶）
taylor_expansion = series(f, x, 0, 5).removeO()  # 展开到4阶
print(f"泰勒级数展开: {taylor_expansion}")

# 将泰勒级数展开转换为可计算的函数
taylor_f = lambda x_val, y_val: float(taylor_expansion.subs({x: x_val, y: y_val}))

# 创建 x, y 网格
x_vals = np.linspace(-2, 2, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# 计算对应的 Z 值（泰勒级数展开的结果）
Z = np.vectorize(taylor_f)(X, Y)

# 创建图形和3D坐标轴
fig = plt.figure(figsize=(12, 6))

# 绘制三维表面图
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('3D Surface Plot of Taylor Series Expansion')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x, y)')

# 绘制等高线图
ax2 = fig.add_subplot(122)
cp = ax2.contour(X, Y, Z, 20, cmap='viridis')
ax2.set_title('Contour Plot of Taylor Series Expansion')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
fig.colorbar(cp, ax=ax2)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义时间区间和基本参数
t = np.linspace(0, 1, 1000)  # 时间从0到1，1000个点
T = 1  # 信号周期
f = 1 / T  # 信号频率

# 方波信号（基本信号）
def square_wave(t):
    return np.sign(np.sin(2 * np.pi * f * t))

# 傅里叶级数展开（近似方波）
def fourier_series(t, N):
    sum = np.zeros_like(t)
    for n in range(1, N+1, 2):  # 只取奇数项 (1, 3, 5, ...)
        sum += (4 / (np.pi * n)) * np.sin(2 * np.pi * n * f * t)
    return sum

# 绘制图形
plt.figure(figsize=(10, 6))

# 绘制原始方波
plt.subplot(2, 1, 1)
plt.plot(t, square_wave(t), label="Square Wave (Original)", color='black')
plt.title("Original Square Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()

# 绘制傅里叶级数展开的近似波形
N_values = [3, 5, 10, 50]  # 展开项的数量
plt.subplot(2, 1, 2)
for N in N_values:
    plt.plot(t, fourier_series(t, N), label=f"N={N}")
plt.title("Fourier Series Approximation")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 定义洛伦兹系统方程
def lorenz(t, state, sigma=10, rho=28, beta=8/3):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# 初始状态
initial_state = [1.0, 1.0, 1.0]  # 初始条件
t_span = (0, 40)  # 时间区间
t_eval = np.linspace(0, 40, 10000)  # 采样点

# 求解洛伦兹系统
solution = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval)

# 提取解
x, y, z = solution.y

# 绘制三维图像
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_title("Lorenz System")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 泊松分布的参数
lambda_ = 4  # 平均事件发生次数（泊松分布的参数）

# 创建x轴值（可能的事件次数）
x = np.arange(0, 15)

# 使用scipy的poisson分布计算PMF
pmf = poisson.pmf(x, lambda_)

# 绘制泊松分布的PMF图
plt.bar(x, pmf, width=0.5, color='skyblue', edgecolor='black')
plt.title('Poisson Distribution PMF')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.xticks(x)
plt.ylim(0, max(pmf) + 0.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 正态分布的参数
mean = 0  # 均值
std_dev = 1  # 标准差

# 创建x轴值（可能的数值范围）
x = np.linspace(-5, 5, 1000)

# 使用scipy的norm分布计算PDF
pdf = norm.pdf(x, mean, std_dev)

# 绘制正态分布的PDF图
plt.plot(x, pdf, color='blue', label=f'Normal Distribution\nmean={mean}, std={std_dev}')
plt.title('Normal Distribution (PDF)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# 卡方分布的参数
df = 5  # 自由度

# 创建x轴值（可能的数值范围）
x = np.linspace(0, 20, 1000)

# 使用scipy的chi2分布计算PDF
pdf = chi2.pdf(x, df)

# 绘制卡方分布的PDF图
plt.plot(x, pdf, color='blue', label=f'Chi-Square Distribution\n(df={df})')
plt.title('Chi-Square Distribution (PDF)')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```
