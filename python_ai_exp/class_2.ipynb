```python
from scipy.stats import norm
```

```python
x = np.linspace(-5, 5, 1000)
pdf_values = norm.pdf(x, loc=0, scale=1)
cdf_values = norm.cdf(x, loc=0, scale=1)

fig, ax1 = plt.subplots()

# 绘制 PDF
ax1.plot(x, pdf_values, 'b-', label='PDF')
ax1.set_xlabel('x')
ax1.set_ylabel('PDF', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 创建第二个y轴，绘制 CDF
ax2 = ax1.twinx()
ax2.plot(x, cdf_values, 'r-', label='CDF')
ax2.set_ylabel('CDF', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# 显示网格
ax1.grid(True)

plt.show()
```

```python
from scipy.optimize import minimize
```

```python
def objective_function(x):
    return (x-3)**2 + 10 *np.sin(x)


x0 = 0

result = minimize(objective_function, x0, method='BFGS')

x = np.linspace(-5, 5, 1000)
y = objective_function(x)
plt.figure()
plt.plot(x,y)
plt.scatter(result.x, objective_function(result.x),color='r',label='Objective function')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv

# Define the order of the Bessel function
v = 0  # You can change this to other orders like 1, 2, etc.

# Define the range of x values
x = np.linspace(0, 20, 400)

# Calculate the Bessel function of the first kind for the given order
y = jv(v, x)

# Plot the Bessel function
plt.plot(x, y, label=f'J_{v}(x)', color='b')
plt.title(f'Bessel Function of the First Kind J_{v}(x)')
plt.xlabel('x')
plt.ylabel(f'J_{v}(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置中文字体
rcParams['font.sans-serif'] = ['SimHei']  # 黑体字体
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 采样频率
fs = 1000  # 1 kHz
# 采样时间
T = 1.0  # s
# 信号持续时间
t = np.linspace(0.0, 1.0, fs, endpoint=False)
# 生成信号：两个正弦波的叠加
f1 = 50.0  # 第一个正弦波的频率 (50 Hz)
f2 = 120.0  # 第二个正弦波的频率 (120 Hz)
y = np.sin(2.0 * np.pi * f1 * t) + 1.0 * np.sin(2.0 * np.pi * f2 * t)

# 计算频谱
yf = fft(y)
xf = fftfreq(len(y), T)[:len(y)//2]

# 绘制频谱图
plt.plot(xf, 2.0/len(y) * np.abs(yf[:len(y)//2]))
plt.grid()
plt.xlabel('频率 (Hz)')
plt.ylabel('振幅')
plt.title('Frequency Spectrum')
plt.show()
```

```python
# 生成一个二维网格数据
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制等高线图
plt.contour(X, Y, Z, levels=10, cmap='RdGy')

# 添加标题和标签
plt.title("等高线图示例")
plt.xlabel("X轴")
plt.ylabel("Y轴")

# 显示图形
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("三维曲面图示例")
ax.set_xlabel("X轴")
ax.set_ylabel("Y轴")
ax.set_zlabel("Z轴")

plt.show()
```
